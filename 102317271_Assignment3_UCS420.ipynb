6. Create a sample dataset (employees.csv) containing information about employees in a company.

```python
data = {
    'Employee_ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Edward'],
    'Department': ['HR', 'IT', 'IT', 'Marketing', 'Sales'],
    'Age': [29, 34, 41, 28, 38],
    'Salary': [50000, 70000, 65000, 55000, 60000],
    'Years_of_Experience': [4, 8, 10, 3, 12],
    'Joining_Date': ['2020-03-15', '2017-07-19', '2013-06-01', '2021-02-10', '2010-11-25'],
    'Gender': ['Female', 'Male', 'Male', 'Female', 'Male'],
    'Bonus': [5000, 7000, 6000, 4500, 5000],
    'Rating': [4.5, 4.0, 3.8, 4.7, 3.5]
}
df = pd.DataFrame(data)
df.to_csv('employees.csv', index=False)
df
```

```python
averagesal = df['Salary'].mean()
print('Average Salary:', averagesal)
```

```python
totalbonus = df['Bonus'].sum()
print('Total Bonus:', totalbonus)
```

```python
minage = df['Age'].min()
print('Youngest Age:', minage)
```

```python
maxrating = df['Rating'].max()
print('Highest Rating:', maxrating)
```

```python
print('Sorted by Salary (Descending):')
df.sort_values(by='Salary', ascending=False)
```

```python
def x(rating):
    if rating >= 4.5:
        return 'Excellent'
    elif rating >= 4.0:
        return 'Good'
    else:
        return 'Average'

df['Performance_Category'] = df['Rating'].apply(x)
print('With Performance Category:')
df
```

```python
missing_values = df.isnull().sum()
print('Missing Values:')
print(missing_values)
```

```python
df.rename(columns={'Employee_ID': 'ID'}, inplace=True)
print('Renamed Employee_ID to ID:')
df
```

```python
print('More than 5 years of experience:')
df[df['Years_of_Experience'] > 5]
```

```python
it = df[df['Department'] == 'IT']
print('IT Department:')
it
```
