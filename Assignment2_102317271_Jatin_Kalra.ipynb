3. WAP to create a list of 100 random numbers between 100 and 900. Count and print
the:
i. All odd numbers
ii. All even numbers
iii. All prime numbers

```python
import random
A = []
for i in range(100):
  A.append(random.randint(100,900))
print("Printing Prime Numbers")
for i in A:
  flag=True
  for j in range(2,i):
    if(i%j==0):
        flag=False
        break
  if(flag):
    print(i,end = " ")
print("\nPrinting Even Numbers")
for i in A:
  if(i%2==0):
    print(i,end = " ")
print("\nPrinting Odd Numbers")
for i in A:
  if(i%2!=0):
    print(i,end = " ")
```

4. Consider the following two sets, A and B, representing scores of two teams in mul∆üple
matches. A = {34, 56, 78, 90} and B = {78, 45, 90, 23}
WAP to perform the following operations using set functions:
i. Find the unique scores achieved by both teams (union of sets).
ii. Iden∆üfy the scores that are common to both teams (intersec∆üon of sets).
iii. Find the scores that are exclusive to each team (symmetric difference).
iv. Check if the scores of team A are a subset of team B, and if team B's scores are
a superset of team A.
v. Remove a specific score ùëã (input by the user) from set A if it exists. If not, print
a message saying it is not present.

```python
A = {34, 56, 78, 90}
B = {78, 45, 90, 23}

uni = A | B
print("Unique scores:", uni)

com = A & B
print("Common scores:", com)

exc = A ^ B
print("Exclusive scores:", exc)

sub = A.issubset(B)
print("A a subset of B?", sub)

sup = B.issuperset(A)
print("B a superset of A?",sup)

X = int(input("Enter the score to remove from A: "))
if X in A:
    A.remove(X)
    print("Set after removal:", A)
else:
    print("Score ",X," is not present in set A.")
```
