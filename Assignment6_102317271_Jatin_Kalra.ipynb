Q1. Ask the user to enter a value (e.g., M) for a mathematical function.
 Generate x values from -10 to 10 using np.linspace().
 Compute y values for:
o y= M⋅x
2
o y=M⋅sin(x)

```python
import numpy as np
import matplotlib.pyplot as plt

M = float(input("Enter the value for M: "))

x = np.linspace(-10, 10, 500)

y1 = M * x**2
y2 = M * np.sin(x)

plt.figure()
plt.plot(x, y1, label="y = M * x^2", color='blue', linestyle='-')
plt.plot(x, y2, label="y = M * sin(x)", color='red', linestyle='--')
plt.legend()
plt.grid(True)
plt.title(f"Graphs of y = M * x^2 and y = M * sin(x) for M = {M}")
plt.show()
```

Q.2 Create a dataset of five subjects and scores.
 Convert it into a Pandas DataFrame.
 Plot the scores using a Seaborn bar plot with:
o Different colors for each bar.
o Annotations on top of each bar.
o Title, axis labels, and grid.

```python
import pandas as pd
import seaborn as sns
data = {
    'Subject': ['English', 'Science', 'Maths', 'Computer', 'Social Studies'],
    'Score': [85, 90, 78, 88, 92]
}

df = pd.DataFrame(data)
bar_plot = sns.barplot(x='Subject', y='Score', data=df, palette='colorblind')

for index, row in df.iterrows():
    bar_plot.text(index, row['Score'], round(row['Score'], 2), color='black', ha="center")

plt.title('Scores by Subject')
plt.xlabel('Subject')
plt.ylabel('Score')
plt.grid(False)

plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().
 Create a 2x2 subplot layout displaying:
o Line plot showing cumulative sum.
o Scatter plot with random noise.
 Customize with titles, axis labels, and grids.

```python
rollno = 102317271
np.random.seed(rollno)
data = np.random.randn(50)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].plot(np.cumsum(data))
axs[0, 0].set_title('Cumulative Sum')
axs[0, 0].set_xlabel('Index')
axs[0, 0].set_ylabel('Cumulative Sum')
axs[0, 0].grid(True)

axs[0, 1].scatter(range(50), data)
axs[0, 1].set_title('Scatter Plot with Random Noise')
axs[0, 1].set_xlabel('Index')
axs[0, 1].set_ylabel('Value')
axs[0, 1].grid(True)

axs[1, 0].axis('off')
axs[1, 1].axis('off')

plt.show()
```

Apply ‘seaborn’ library to do the following.
1. Read Total profit of all months and show it using a line plot.
2. Read all product sales data and show it using a multiline plot.
3. Plot bar chart for all the features/attributes.

```python
url = 'https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv?raw=true'
data = pd.read_csv(url)

plt.figure(figsize=(10, 6))
sns.lineplot(x=data['month_number'], y=data['total_profit'])
plt.title('Total Profit of All Months')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.lineplot(x=data['month_number'], y=data['facecream'], label='Face Cream Sales')
sns.lineplot(x=data['month_number'], y=data['facewash'], label='Face Wash Sales')
sns.lineplot(x=data['month_number'], y=data['toothpaste'], label='Toothpaste Sales')
sns.lineplot(x=data['month_number'], y=data['bathingsoap'], label='Bathing Soap Sales')
sns.lineplot(x=data['month_number'], y=data['shampoo'], label='Shampoo Sales')
sns.lineplot(x=data['month_number'], y=data['moisturizer'], label='Moisturizer Sales')
plt.title('Product Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()

data.sum().plot(kind='bar', figsize=(12, 8))
plt.title('Total Sales and Profit of All Products')
plt.xlabel('Product/Attribute')
plt.ylabel('Total Value')
plt.grid(True)
plt.show()
```
