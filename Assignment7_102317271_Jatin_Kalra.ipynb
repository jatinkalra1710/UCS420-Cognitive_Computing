```python
import numpy as np
import pandas as pd

np.random.seed(102317271)

sales_data = np.random.randint(1000, 5001, size=(12, 4))

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
df_sales = pd.DataFrame(sales_data, columns=categories, index=months)

df_sales
```

```python
rows = df_sales.head()
print('Top 5 Rows')
print(rows)
print('\n')
stats = df_sales.describe()
print(stats)
print('\n')

totalcategory = df_sales.sum()
print(totalcategory)
print('\n')
totalmonth = df_sales.sum(axis=1)
print(totalmonth)
print('\n')

averagesales = df_sales.pct_change().mean()
print(averagesales)

df_sales['Total Sales'] = df_sales.sum(axis=1)
df_sales['Growth Rate'] = df_sales['Total Sales'].pct_change() * 100
print(df_sales)

roll_number = 102317271

if roll_number % 2 == 0:
    df_sales['Electronics'] = df_sales['Electronics'] * 0.9
else:
    df_sales['Clothing'] = df_sales['Clothing'] * 0.85

df_sales
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
for category in categories:
    plt.plot(df_sales.index, df_sales[category], marker='o', label=category)
plt.title('Monthly Sales Trends for Each Category')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_sales[categories])
plt.title('Sales Distribution for Each Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
import numpy as np

array = np.array([[1, -2, 3], [-4, 5, -6]])

abs_array = np.abs(array)

flattened_array = array.flatten()
percentiles_flattened = np.percentile(flattened_array, [25, 50, 75])
percentiles_columns = np.percentile(array, [25, 50, 75], axis=0)
percentiles_rows = np.percentile(array, [25, 50, 75], axis=1)

mean_flattened = np.mean(flattened_array)
mean_columns = np.mean(array, axis=0)
mean_rows = np.mean(array, axis=1)

median_flattened = np.median(flattened_array)
median_columns = np.median(array, axis=0)
median_rows = np.median(array, axis=1)

std_flattened = np.std(flattened_array)
std_columns = np.std(array, axis=0)
std_rows = np.std(array, axis=1)

print(abs_array,
 percentiles_flattened,
 percentiles_columns,
 percentiles_rows,
 mean_flattened,
 mean_columns,
 mean_rows,
 median_flattened,
 median_columns,
 median_rows,
 std_flattened,
 std_columns,
 std_rows)
```

```python
def swap_elements(lst, i, j):
    temp = lst[i]
    lst[i] = lst[j]
    lst[j] = temp
list1 = [1, 2, 3, 4, 5]
print("Original list:", list1)

swap_elements(list1, 1, 3)
print("List after swapping elements at indices 1 and 3:", list1)
```

```python
def swap(s, e1, e2):

    lst1 = list(s)

    i1 = lst1.index(e1)
    i2 = lst1.index(e2)

    temp = lst1[i1]
    lst1[i1] = lst1[i2]
    lst1[i2] = temp

    swapped_set = set(lst1)
    return swapped_set

set1 = {1, 2, 3, 4, 5}
print("Original set:", set1)

swapped_set = swap(set1, 2, 4)
print("Set after swapping elements 2 and 4:", set1)
```
