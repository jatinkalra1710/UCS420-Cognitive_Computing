b) Find the most frequent value and their indice(s) in the following arrays:
i. x = np.array([1,2,3,4,5,1,2,1,1,1])
ii. y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ])

```python
x = np.array([1,2,3,4,5,1,2,1,1,1])
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])
a,b = np.unique(x, return_counts=True)
elementx = (a[np.argmax(b)])
indexx = np.where(x == elementx)
print("Max occurrence is of : ",elementx)
print("Indices : ",indexx[0])
```

```python
c,d = np.unique(y, return_counts=True)
elementy = (c[np.argmax(d)])
indexy = np.where(y == elementy)
print("Max occurrence is of : ",elementy)
print("Indices : ",indexy[0])
```

Q.4 Write program to create an 1-D NumPy array named YourName with evenly
spaced 25 numbers from 10 to 100 using linspace(). Print the dimensions of the array,
shape, total elements, the data type of each element and total number of bytes consumed
by the array. Find the transpose of this array using reshape() attribute. Can we do the same
with T attribute?

```python
jatinkalra = np.linspace(10,100,25)
print("Dimension : ",jatinkalra.ndim)
print("Shape : ",jatinkalra.shape)
print("Total Elements : ",jatinkalra.size)
print("Data Type : ",jatinkalra.dtype)
print("Total Bytes : ",jatinkalra.nbytes)
print("Transpose using reshape :",jatinkalra.reshape(25,1))
print("Transpose using T : ",jatinkalra.T)
```

Q5. Create a 2-D Array of three rows and four columns, named ucs420_yourname
with following values â€“ 10, 20, 30, 40, 50, 60, 70, 80, 90, 15, 20, 35. Compute the mean,
median, max, min, unique elements. Reshape the array to four rows and three columns and
name it as reshaped_ ucs420_yourname. Resize the array to two rows and three
columns and name it as resized_ ucs420_yourname

```python
ucs420_jatinkalra = np.array([10,20,30,40,50,60,70,80,90,15,20,35])
print("Mean : ",np.mean(ucs420_jatinkalra))
print("Median : ",np.median(ucs420_jatinkalra))
print("Max : ",np.max(ucs420_jatinkalra))
print("Min : ",np.min(ucs420_jatinkalra))
print("Unique : ",np.unique(ucs420_jatinkalra))
reshaped_ucs420_jatinkalra = ucs420_jatinkalra.reshape(4,3)
print("Reshaped : ",reshaped_ucs420_jatinkalra)
resized_ucs420_jatinkalra = np.resize(ucs420_jatinkalra,(2,3))
print("Resized : ",resized_ucs420_jatinkalra)
```
